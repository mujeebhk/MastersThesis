```python
import os
import cv2
def load_images_from_folder(folder):
    images = []
    COUNTER=0
    image_count=len(os.listdir(folder))
    train_count=int(((image_count*80)/100)/2)
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename),cv2.IMREAD_GRAYSCALE)
        if('jpg' in filename):
            COUNTER=COUNTER+1
            try:
                if(COUNTER>train_count):
                    file=open('/Users/mohammedmujeeb/Google Drive/Masters Thesis/VocalCordCyst.v1.darknet/combined/test.txt','a')
                else:
                    file=open('/Users/mohammedmujeeb/Google Drive/Masters Thesis/VocalCordCyst.v1.darknet/combined/train.txt','a')
                file.write('data/obj/'+filename+'\n')
                file.close
#                 print(os.path.dirname(os.path.abspath('trainDDD.txt')))
    #             img = cv2.resize(img, (50, 50))
                print('ok')
            except Exception as e:
                print('skipping file', filename,e)
                COUNTER=COUNTER-1
       
#         cv2.imwrite('/Users/jameel/Desktop/machine learning/jupyter/HAAR/02 image factory/pos/'+str(COUNTER)+'.jpeg',img)
    print('done processing image ', COUNTER)

load_images_from_folder('/Users/mohammedmujeeb/Google Drive/Masters Thesis/VocalCordCyst.v1.darknet/combined')
```
